```python
# Enriching the Data
# One of the most important aspects of a job would be the salary information. So we will try to obtain the minimum salary and the maximum salary from the salary column.
```

```python
import re
```

```python
df ['Salary_Estimate'] = df['J_Salary'].apply(lambda x: re.sub('[^0-9-]', '', x))
```

```python
df.head()
```

```python
# df['Min_Salary'] = df['Salary_Estimate'].apply(lambda x: x.split('-')[0])
```

```python
# df.head()
```

```python
# new col. with min. salary

df.insert(2, 'Min_Salary', df['Salary_Estimate'].apply(lambda x: float(x.split('-')[0])))
```

```python
df.head()
```

```python
# new col. with Max. salary

df.insert(3, 'Max_Salary', df['Salary_Estimate'].apply(lambda x: float(x.split('-')[1])))
```

```python
df[['Min_Salary', 'Max_Salary']].describe()
```

```python
# Binning
# Now since we have extracted the min and max salaries, we need to classify the jobs into 3 categories - Low paying jobs, medium paying jobs and high paying jobs. Let us see how we can accomplish this.
```

```python
df['Max_Sal_Range'] = pd.qcut(df['Max_Salary'], q=3, labels=['L', 'M', 'H'] )
```
